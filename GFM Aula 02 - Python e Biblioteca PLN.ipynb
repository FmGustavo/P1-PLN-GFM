## Exemplo 06 - Laços de repetição

```python
# while (Condição): (Instruções a serem repetidas)

# for (Contador) in range(Contagem): (Instruções a serem repetidas)

i = 0

# variavel simples- armazena apenas um valor
# variavel composta - armazena vários valores
  # lista - mutáveis: [a, b, c, 45, [ad]]
  # tupla - imutáveis: (a, b, c, d, 45, asd, asd, (asdcfsd))
  # conjuntos - mutável mas sem repetição {valor1, valor2, valor3,...}

number = [] # Inicialização de uma variável composta do tipo mista

while (i < 6):
  print()
  print(f"Estou indo de 0 até {i}")
  number.append(i)
  i = i + 1
  print(f"o valor da lista é {number}")
  # Aqui esta dentro da repetição

# Aqui esta fora da repetição

print()
print(f"Vamos experimentar o for")

for i in range(0,6):
  print()
  print(f"Estou indo de 0 até {i}")
  number.append(i)
  print(f"O valos da lista é {number}")
  # Aqui está dentro da repetição

print()
print(f"Vamos experimentar o for com strings.")
text = "Aula de sexta feira"

for i in text:
  print (f"O contador esta com a letra {i}.")
```

## Exercício 1

### Elaborar um programa que calcule e apresente o valor do volume de uma caixa retangular, utilizando a fórmula VOLUME = COMPRIMENTO * LARGURA * ALTURA

```python
# Adquirir as medidas da caixa retangular
Comprimento = int(input("Qual é o comprimento da caixa? "))
print ()

Largura = int(input("Qual é a largura da caixa? "))
print ()

Altura = int(input("Qual é a altura da caixa? "))
print ()

# Calcular o volume da caixa
Volume = Comprimento * Largura * Altura

# Informar o volume da caixa
print(f"O volume da caixa é: {Volume}")
```

## Exercício 2

### Criar um programa que leia um texto e conte o número de palavras

```python
# Função para contar o número de palavras em um texto
def contar_palavras(texto):
    # Dividir o texto em palavras usando o espaço como delimitador
    palavras = texto.split()
    # Retornar o número de palavras
    return len(palavras)

# Ler o texto do usuário
texto = input("Digite o texto: ")
print ()

# Contar o número de palavras
numero_de_palavras = contar_palavras(texto)

# Exibir o resultado
print(f'O número de palavras no texto é: {numero_de_palavras}')
```

## Exercício 3

### Buscar e substituir palavras em um texto

```python
# Função para buscar e substituir palavras
def buscar_substituir(texto, palavra_antiga, palavra_nova):

    # Substitui todas as ocorrências de palavra_antiga por palavra_nova
    texto_substituido = texto.replace(palavra_antiga, palavra_nova)
    return texto_substituido

# Ler o texto do usuário
texto = input("Digite o texto: ")
print ()

# Ler a palavra que será substituída e a nova palavra
palavra_antiga = input("Digite a palavra a ser substituída: ")
print ()
palavra_nova = input("Digite a nova palavra: ")
print ()

# Chamar a função para realizar a substituição
texto_modificado = buscar_substituir(texto, palavra_antiga, palavra_nova)

# Exibir o texto modificado
print(f"Texto modificado: {texto_modificado}")
```

## Exercício 4

### Separar frases usando split() e iterar sobre elas

```python
# Função para separar e iterar sobre as frases
def separar_frases(texto):

    # Dividir o texto com base nos delimitadores de frases (., ?, !)
    # Adicionamos um espaço após o ponto para garantir que as frases não fiquem coladas
    frases = texto.split('. ')

    # Iterar sobre as frases e exibir
    for i, frase in enumerate(frases):
        print(f"Frase {i+1}: {frase.strip()}")

# Ler o texto do usuário
print ()
texto = input("Digite o texto: ")

# Chamar a função para separar e iterar sobre as frases
separar_frases(texto)
```

## Exercício 5

### Ler três valores inteiros representados pelas variáveis A, B e C e apresentar os valores lidos dispostos em ordem crescente

```python
# Ler os três valores inteiros
A = int(input("Digite o valor de A: "))
print ()
B = int(input("Digite o valor de B: "))
print ()
C = int(input("Digite o valor de C: "))
print ()

# Colocar os valores em uma lista
valores = [A, B, C]

# Ordenar os valores em ordem crescente
valores_ordenados = sorted(valores)

# Exibir os valores ordenados
print(f"Os valores em ordem crescente são: {valores_ordenados[0]}, {valores_ordenados[1]}, {valores_ordenados[2]}")
```

## Exercício 6

### Fazer a leitura de quatro valores numéricos inteiros representados pelas variáveis A, B, C e D. Apresentar apenas os valores que sejam divisíveis por 2 e 3

```python
# Ler os quatro valores inteiros
A = int(input("Digite o valor de A: "))
print ()
B = int(input("Digite o valor de B: "))
print ()
C = int(input("Digite o valor de C: "))
print ()
D = int(input("Digite o valor de D: "))
print ()

# Verificar e apresentar os valores divisíveis por 2 e 3
valores = [A, B, C, D]

for valor in valores:
    if valor % 2 == 0 and valor % 3 == 0:
        print(f"O valor {valor} é divisível por 2 e 3.")
    else:
        print(f"O valor {valor} não é divisível por 2 e 3")
```
